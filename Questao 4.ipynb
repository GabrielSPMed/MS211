```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
#matriz criada, coluna final preenchida com 0.01, restante com 0
a = 100
matrix = [[0 for x in range(a+1)]for y in range (a)]
for i in range (a):
    matrix[i][100]=0.01
```

```python
#diagonal principal preenchida com 550.05
for i in range(100):
    matrix[i][i]=550.05
```

```python
#diagonal preenchida com-325
for i in range(99):
    matrix[i+1][i] = -325
```

```python
#diagonal preenchida com -175
for i in range(99):
    matrix[i][i+1] = -175
```

```python
#valores colocados na coluna final
matrix[34][100] = 0.2
matrix[39][100] = 0.5
matrix[44][100] = 0.85
matrix[49][100] = 0.35
matrix[54][100] = 0.15
```

```python
#metodo da eliminacao de gauss-jordan
for linha in range (100):
    #simplificando as linhas
    for coluna in range(101):
        if (linha!= coluna):
            matrix[linha][coluna] = matrix[linha][coluna]/matrix[linha][linha]
    matrix[linha][linha]=1
    #subtracao das linhas inferiores e superiores a fim de gerar a matriz identidade
    for linha_subtraida in range(100):
        if linha_subtraida != linha:
            if matrix[linha_subtraida][linha] != 0:
                multiplicador = -1 * matrix[linha_subtraida][linha]
                for coluna_subtracao in range (linha, 101):
                    matrix[linha_subtraida][coluna_subtracao] += multiplicador * matrix[linha][coluna_subtracao]         
```

```python
vect= [matrix[i][100] for i in range (100)]
```

```python
x = [i + 1 for i in range(100)]
plt.plot(x, vect, 'bo')
plt.ylabel('Valor obtido')
plt.xlabel('Posicao no vetor x')
plt.show()
```
